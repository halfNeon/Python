```python
data = scaled_df.drop('country',axis = 1)
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    ssd.append(km.inertia_) 
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmean = KMeans(n_clusters=3)
kmean.fit(data)
```

```python
pred = kmean.labels_
print(pred)
```
